###Write a code that that prduces  list of all partititions of the set of $[n]=\{ 1,2,3,4\cdots , n-1, n\}$ with $0 \le n \le 65$ and also list all the partitions for wich the sun of elemst of tha parttionm es is the sme constann for all.

```python
def binomial_coefficient(n, k):
    if k == 0 or k == n:
        return 1
    if k > n - k:            #usar la simetr√≠a para ahorrar pasos
        k = n - k
    result = 1
    for i in range(1, k + 1):
        result = result * (n - i + 1) // i
    return result

def bell_number(n):
    if n == 0:
        return 1
    return sum(binomial_coefficient(n-1, k) * bell_number(k) for k in range(n))

for i in range(10):
    print(f"Bell number for n={i} is: {bell_number(i)}")
```

```python
def generate_partitions(set_elements):
    """Generar todas las particiones de un conjunto."""
    if not set_elements:
        yield []
        return
    first, rest = set_elements[0], set_elements[1:]
    for smaller in generate_partitions(rest):
        yield [[first]] + smaller
        for idx, subset in enumerate(smaller):
            yield smaller[:idx] + [[first] + subset] + smaller[idx+1:]

def constant_sum_partitions(n):
    """Filtrar particiones donde todos los subconjuntos tienen la misma suma."""
    for partition in generate_partitions(list(range(1, n+1))):
        sums = [sum(subset) for subset in partition]
        if len(set(sums)) == 1:
            yield partition

n = 7
print("All Partitions:")
for partition in generate_partitions(list(range(1, n+1))):
    print(partition)
print("\nConstant Sum Partitions:")
for partition in constant_sum_partitions(n):
    print(partition)
```

# Write a code that produces the partitiosn with all the parts havig same sum value, identify the formla for $n < 16$.

```python
def all_subsets(nums):
    """
    Genera todos los subconjuntos de una lista de n√∫meros.

    Args:
    - nums (list): Lista de n√∫meros.

    Returns:
    - list: Lista de todos los subconjuntos.
    """
    if len(nums) == 0:
        return [[]]

    subsets = all_subsets(nums[:-1])
    next_elem = nums[-1]

    return subsets + [item + [next_elem] for item in subsets]

def partitions_with_equal_sum(nums):
    """
    Encuentra particiones donde todos los subconjuntos tienen la misma suma.

    Args:
    - nums (list): Lista de n√∫meros.

    Returns:
    - list: Lista de particiones.
    """
    # Generar todos los subconjuntos posibles
    subsets = all_subsets(nums)
    sum_dict = {}

    # Categorizar subconjuntos por suma
    for subset in subsets:
        subset_sum = sum(subset)
        if subset_sum in sum_dict:
            sum_dict[subset_sum].append(subset)
        else:
            sum_dict[subset_sum] = [subset]

    # Filtrar aquellos con m√°s de un subconjunto
    partitions = [value for key, value in sum_dict.items() if len(value) > 1]

    return partitions

nums = [i for i in range(1, 16)]
result = partitions_with_equal_sum(nums)
for partition in result:
    print(partition)
```
